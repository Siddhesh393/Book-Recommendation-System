## Visualization

```python
plt.figure(figsize=(8,5))
sns.countplot(x='Book-Rating', data=rating, palette='coolwarm')
plt.title("Distribution of Book Ratings")
plt.xlabel("Rating")
plt.ylabel("Count")
plt.show()
```

```python
plt.figure(figsize=(8,5))
sns.histplot(user['Age'], bins=20, kde=True)
plt.title("Distribution of User Age")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()
```

```python
top_books = rating.groupby('ISBN')['Book-Rating'].mean().sort_values(ascending=False).head(10)
print(top_books)
```

```python
active_users = rating['User-ID'].value_counts().head(10)
print(active_users)
```

```python
top_locations = user['Location'].value_counts().head(10)
print(top_locations)
```

```python
merged_df = rating.merge(book, on='ISBN').merge(user, on='User-ID')
merged_df.head()
```

```python
popular_books = merged_df.groupby('Book-Title').size().sort_values(ascending=False).head(10)
popular_books.plot(kind='barh', figsize=(8,5), color='skyblue')
plt.title("Top 10 Most Rated Books")
plt.xlabel("Number of Ratings")
plt.ylabel("Book Title")
plt.show()
```

```python
import requests
from PIL import Image
from io import BytesIO
import matplotlib.pyplot as plt

url = book[book['Book-Title'] == 'The Wild Baron']['Image-URL-L'].values[0] 
url = book[book['Book-Title'] == 'Willful Marriage (Silhouette Desire, 1024)']['Image-URL-L'].values[0] 
# Set headers to mimic a browser
headers = {
    "User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/141.0.0.0 Safari/537.36"
}

response = requests.get(url, headers=headers)

# Check if the content is an image
if response.status_code == 200 and 'image' in response.headers['Content-Type']:
    img = Image.open(BytesIO(response.content))
    plt.imshow(img)
    plt.axis('off')
    plt.show()
else:
    print("Cannot display image. URL may be blocked or not an image.")
```
